```python
#paths are dynamically present
# Step 1: Collect User Input
video_idea = "Generate a text on coolest places on earth for a 2-minute video"
video_duration = 120  # in seconds

# Step 2: Generate Script
import os
from groq import Groq

client = Groq(
    api_key="Your_API_Key",
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"""Generate a script for a {video_duration//60}-minute video on: {video_idea}
            The output should be in the example format: **Title:** "Explore the Coolest Places on Earth"

**Intro:** (0:00 - 0:10)

(Upbeat background music starts playing. The camera pans over a stunning natural landscape, gradually revealing the video's title)

Voiceover: "Are you ready to explore the most breathtaking, awe-inspiring, and coolest places on earth? From icy glaciers to scorching hot deserts, we've got you covered. Join us on this thrilling journey to the coolest destinations on our planet!"

**Section 1: Icy Wonders** (0:10 - 0:40)

(Cut to footage of glaciers, ice caves, and snow-capped mountains)

Voiceover: "From the frozen tundras of Antarctica to the majestic glaciers of Iceland, the coolest places on earth are awe-inspiring and humbling. Witness the ethereal beauty of ice caves, hike to the tops of towering mountains, and marvel at the serenity of frozen landscapes."

**Section 2: Scorching Deserts** (0:40 - 1:10)

(Cut to footage of sand dunes, mirages, and oases in the desert)

Voiceover: "But the coolest places on earth aren't just about snow and ice. Get ready to feel like you're in a different world in the scorching hot deserts of Africa, Asia, and the Americas. Explore towering sand dunes, chase mirages, and discover the secrets of these mystical landscapes."

**Section 3: Bizarre Landscapes** (1:10 - 1:40)

(Cut to footage of unique rock formations, crystal caves, and natural wonders)

Voiceover: "And then there are the places that defy explanation. Join us on a journey to the most bizarre and fascinating landscapes on earth, from eerie rock formations to crystal caves and natural wonders that will leave you speechless."

**Conclusion:** (1:40 - 2:00)

(Cut back to the video's title screen)

Voiceover: "From the coldest to the hottest, and from the most bizarre to the most breathtaking, we hope you've enjoyed this journey to the coolest places on earth. Remember to explore, to discover, and to never stop being amazed by the natural wonders of our incredible planet."

**Outro:** (2:00)

(Closing shot of the video's title screen)

Voiceover: "Subscribe to our channel for more epic adventures and breathtaking destinations. See you in the next video!"

(Closing shot of the video)""",
        }
    ],
    model="llama3-8b-8192",
)

script = chat_completion.choices[0].message.content
print(script)

# Step 3: Divide Script into Chunks for voice over part only
import re

# Define the keywords to identify sections
keywords = ["Intro", "Section 1", "Section 2", "Section 3", "Conclusion", "Outro"]

# Split the script into lines
lines = script.split('\n')

# Initialize variables
chunks = []
current_chunk = ""
current_section = "unknown_section"  # Initialize with a default value

# Iterate through lines and create chunks based on keywords
for line in lines:
    if any(keyword in line for keyword in keywords):
        if current_chunk:
            chunks.append((current_section, current_chunk.strip()))
        current_section = line
        current_chunk = line + "\n"
    else:
        current_chunk += line + "\n"

# Add the last chunk if any lines are left
if current_chunk:
    chunks.append((current_section, current_chunk.strip()))

# Extract voiceover text from chunks
voiceover_chunks = []
for section, chunk in chunks:
    voiceover_match = re.search(r'Voiceover: "(.*?)"', chunk, re.DOTALL)
    if voiceover_match:
        voiceover_text = voiceover_match.group(1)
        voiceover_chunks.append((section, voiceover_text))

# Print the voiceover chunks
for section, voiceover_text in voiceover_chunks:
    print(f"Section: {section}\nVoiceover:\n{voiceover_text}\n")

# Step 4: Generate Media Prompts with Hugging Face
from gradio_client import Client
from PIL import Image
import shutil

# Initialize the Hugging Face client with the API key
client = Client("black-forest-labs/FLUX.1-schnell", hf_token="Your_HF_Token")

media_files = []
for section, chunk in chunks:
    # Ensure section is not None
    if section is None:
        section = "unknown_section"

    # Create a prompt for the Hugging Face model
    prompt = f"""Generate an image for the following section of a video script: {section}\n\n{chunk}
               Images should not have any text written over them.
               Output:- Image that suits description without any Text added"""

    # Generate image using Hugging Face model
    try:
        result = client.predict(
            prompt=prompt,
            seed=0,
            randomize_seed=True,
            width=1920,  # Set width to 1920 for 16:9 aspect ratio
            height=1080,  # Set height to 1080 for 16:9 aspect ratio
            num_inference_steps=2,  # Reduce the number of inference steps to reduce GPU usage
            api_name="/infer"
        )

        # Process the generated image
        if result:
            image_path = result[0]
            # Copy the image from the local path to the desired location
            img_path = os.path.join("/content", f"{section.replace(' ', '_').lower()}.png")
            shutil.copy(image_path, img_path)
            media_files.append((section, img_path))
        else:
            print(f"Error: Image generation failed for section {section}")
    except Exception as e:
        print(f"Error: {e} for section {section}")

print(media_files)

# Step 5: Text-to-Speech for voice over part only
from gtts import gTTS

audio_files = []
for section, voiceover_text in voiceover_chunks:
    # Ensure section is not None
    if section is None:
        section = "unknown_section"

    tts = gTTS(text=voiceover_text, lang='en')
    audio_file = os.path.join("/content", f"audio_{section.replace(' ', '_').lower()}.mp3")
    tts.save(audio_file)
    audio_files.append((section, audio_file))

print(audio_files)

# Step 6: Compile Media
import moviepy.editor as mp

# Dynamically generated image paths
image_paths = [img_path for _, img_path in media_files]

# Dynamically generated audio files
audio_files = [audio_file for _, audio_file in audio_files]

image_paths.sort()
audio_files.sort()

# Print the contents of image_paths and audio_files for debugging
print("Image paths:", image_paths)
print("Audio files:", audio_files)

video_clips = []
for image_path, audio_file in zip(image_paths, audio_files):
    print(f"Processing image: {image_path}")
    print(f"Audio file: {audio_file}")
    try:
        video_clip = mp.ImageClip(image_path).set_duration(24).set_fps(24)
        audio_clip = mp.AudioFileClip(audio_file)
        video_clip = video_clip.set_audio(audio_clip)
        video_clips.append(video_clip)
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")

print("Video clips:", video_clips)

# Step 7: Edit Video
from moviepy.video.fx.all import fadein, fadeout

background_music_path = os.path.join("/content", "background_music.mp3")

if video_clips:
    final_video = mp.concatenate_videoclips(video_clips, method="compose")
    final_video = fadein(final_video, 1).fadeout(1)

    if os.path.exists(background_music_path):
        final_video = final_video.set_audio(mp.AudioFileClip(background_music_path))
    else:
        print(f"Background music file '{background_music_path}' not found. Skipping background music.")
else:
    print("No video clips to concatenate.")

print("Final video created.")

# Step 8: Export Video
if video_clips:
    final_video.write_videofile(os.path.join("/content", "final_video.mp4"), codec="libx264", audio_codec="aac", fps=24)
    print("Video generation complete!")
else:
    print("No video clips to export.")
```

```python
#dynamic section part without section1 #overlapping only 2 min generated
# Step 1: Collect User Input
video_idea = "Real-World Applications of Generative AI. Explained in 5 Minutes with visuals and voice over clearly describing for the suitable duration"
video_duration = 300  # in seconds

# Step 2: Generate Script
import os
from groq import Groq

client = Groq(
    api_key="Your_API_Key",
)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": f"""
Generate a detailed script for a {video_duration // 60}-minute video titled: "{video_idea}".
The script should be designed for voiceover, ensuring that the content fills the selected video duration. Aim for approximately **2.5 to 3 spoken words per second** to match the pacing of the voiceover with the video length.
Visual section should ask to generate a single image based on the topic relevant to the section.

### Script Example Format:

**Title:** "{video_idea}"

**Intro:** (0:00 - 0:10)
[Visuals: Generate an introductory image related to the topic.]

Voiceover: "Welcome to our {video_idea} video, where we’ll explore [brief overview of the video’s content]. Stay with us as we dive into this fascinating topic, providing insights and deep knowledge."

**Main Content:** (0:10 - {video_duration - 10} seconds)
[Visuals: Generate relevant images for the main content.]

Voiceover: "[Detailed explanation of the main content.] This part of the journey helps us understand [specific topic points], ensuring you gain a clear understanding of [related concepts]."

**Conclusion:** ({video_duration - 10} - {video_duration} seconds)
[Visuals: Generate a closing image that ties the video’s theme together.]

Voiceover: "Thank you for watching this in-depth exploration of {video_idea}. We hope you’ve gained valuable insights. Don’t forget to like, subscribe, and check out our other videos for more great content!"

### Guidelines:
1. Ensure the script’s length is dynamically adjusted to match the user-defined video duration, while maintaining a clear, engaging voiceover flow.
2. Keep the pacing at **2.5 to 3 spoken words per second** to align with the video’s total duration.
3. Include sufficient content in each section to maintain engagement throughout the video.
4. Adapt the script sections to the video length, ensuring there is detailed, informative content for each segment.
5. Maintain an accessible, clear tone suitable for voiceover delivery.

"""
        }
    ],
    model="llama3-8b-8192",
)

script = chat_completion.choices[0].message.content
print(script)

# Step 3: Divide Script into Chunks for voice over part only
import re

# Define the keywords to identify sections
keywords = ["Intro", "Main Content", "Conclusion"]

# Split the script into lines
lines = script.split('\n')

# Initialize variables
chunks = []
current_chunk = ""
current_section = "unknown_section"  # Initialize with a default value

# Iterate through lines and create chunks based on keywords
for line in lines:
    if any(keyword in line for keyword in keywords):
        if current_chunk:
            chunks.append((current_section, current_chunk.strip()))
        current_section = line
        current_chunk = line + "\n"
    else:
        current_chunk += line + "\n"

# Add the last chunk if any lines are left
if current_chunk:
    chunks.append((current_section, current_chunk.strip()))

# Extract voiceover text from chunks
voiceover_chunks = []
for section, chunk in chunks:
    voiceover_match = re.search(r'Voiceover: "(.*?)"', chunk, re.DOTALL)
    if voiceover_match:
        voiceover_text = voiceover_match.group(1)
        voiceover_chunks.append((section, voiceover_text))

# Print the voiceover chunks
for section, voiceover_text in voiceover_chunks:
    print(f"Section: {section}\nVoiceover:\n{voiceover_text}\n")

# Step 4: Generate Media Prompts with DALL-E
from openai import OpenAI
from PIL import Image
import requests
from io import BytesIO

opclient = OpenAI(api_key="Your_API_Key")

media_files = []
for section, chunk in chunks:
    # Ensure section is not None
    if section is None or section == "unknown_section":
        continue

    # Create a prompt for the DALL-E model
    prompt = f"Generate an image for the following section of a video script: {section}\n\n{chunk}"

    # Generate image using DALL-E model
    try:
        response = opclient.images.generate(
            model="dall-e-3",
            prompt=prompt,
            n=1,
            size="1792x1024"
        )
        image_url = response.data[0].url
        image_response = requests.get(image_url)
        image = Image.open(BytesIO(image_response.content))
        img_path = os.path.join("/content", f"{section.replace(' ', '_').lower()}.png")
        image.save(img_path)
        media_files.append((section, img_path))
    except Exception as e:
        print(f"Error generating image for section {section}: {e}")

print(media_files)

# Step 5: Text-to-Speech for voice over part only
from gtts import gTTS

audio_files = []
for section, voiceover_text in voiceover_chunks:
    # Ensure section is not None
    if section is None or section == "unknown_section":
        continue

    tts = gTTS(text=voiceover_text, lang='en')
    audio_file = os.path.join("/content", f"audio_{section.replace(' ', '_').lower()}.mp3")
    tts.save(audio_file)
    audio_files.append((section, audio_file))

print(audio_files)

# Step 6: Compile Media
import moviepy.editor as mp

# Dynamically generated image paths and audio files
image_paths = [img_path for section, img_path in media_files]
audio_paths = [audio_file for section, audio_file in audio_files]

# Ensure the order of image_paths and audio_files is consistent with the sections
sections_order = [section for section, _ in media_files]

# Create dictionaries for quick lookup
image_dict = {section: img_path for section, img_path in media_files}
audio_dict = {section: audio_file for section, audio_file in audio_files}

# Reorder paths based on the original sections order
image_paths = [image_dict[section] for section in sections_order]
audio_paths = [audio_dict[section] for section in sections_order]

# Print the contents of image_paths and audio_files for debugging
print("Image paths:", image_paths)
print("Audio files:", audio_paths)

video_clips = []
for image_path, audio_file in zip(image_paths, audio_paths):
    print(f"Processing image: {image_path}")
    print(f"Audio file: {audio_file}")
    try:
        video_clip = mp.ImageClip(image_path).set_duration(24).set_fps(24)
        audio_clip = mp.AudioFileClip(audio_file)
        video_clip = video_clip.set_audio(audio_clip)
        video_clips.append(video_clip)
    except Exception as e:
        print(f"Error processing image {image_path}: {e}")

print("Video clips:", video_clips)

# Step 7: Edit Video
from moviepy.video.fx.all import fadein, fadeout

background_music_path = os.path.join("/content", "background_music.mp3")

if video_clips:
    final_video = mp.concatenate_videoclips(video_clips, method="compose")
    final_video = fadein(final_video, 1).fadeout(1)

    if os.path.exists(background_music_path):
        final_video = final_video.set_audio(mp.AudioFileClip(background_music_path))
    else:
        print(f"Background music file '{background_music_path}' not found. Skipping background music.")
else:
    print("No video clips to concatenate.")

print("Final video created.")

# Step 8: Export Video
if video_clips:
    final_video.write_videofile(os.path.join("/content", "final_video.mp4"), codec="libx264", audio_codec="aac", fps=24)
    print("Video generation complete!")
else:
    print("No video clips to export.")
```
